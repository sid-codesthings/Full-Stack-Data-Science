```python
# Stack - LIFO
# Queue - FIFO

# Implement Stack using Queue .

# push()
# pop()
# isEmpty()

# enQueue()
# deQueue()

# so when we push in stack like s.push(5) its doing q.enQueue(5)
# internally.

```

```python
class Node :
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    
    def getData(self):
        return self.data
    
    def setData(self,data):
        self.data = data
        
    def setNext(self,next):
        self.next = next
        
    def getNext(self):
        return self.next
```

```python
class QueueLL:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def isEmpty(self):
        return self.head == None # or self.tail == None
    
    # isFull function not required cz its LL so dynamic sizing
    # here.
    
    def enQueue(self,data):
        node = Node(data)
        
        if self.isEmpty():
            self.head = node
            self.tail = node
        else:
            self.tail.setNext(node)
            self.tail = self.tail.getNext()
        self.size +=1
        
    def deQueue(self):
        if self.isEmpty():
            print("Queue underflow")
            return
        data = self.head.getData()
        
        # if its the only element
        if self.head == self.tail :
            self.head = None
            self.tail = None # because after dequeue, queue will be empty
            
        else:
            self.head = self.head.getNext()
        self.size -=1
        return data
    
    def traverse(self):
        temp = self.head
        while(temp):
            print(temp.getData(),end="->")
            temp = temp.getNext()
        print()
```

```python
# Implement Stack using Queue

class StackQ:
    def __init__(self):
        self.q1 = QueueLL() #it will act as the main queue
        self.q2 = QueueLL()
        
    def isEmpty(self):
        return self.q1.isEmpty()
        
    def push(self,data):
        self.q1.enQueue(data)
        
    def pop(self):
        if self.isEmpty():
            print("Stack underflow error")
        
        # we need to pick the last element added
        data = None
        while not self.q1.isEmpty() :
            data = self.q1.deQueue()
            if(not self.q1.isEmpty()):
                self.q2.enQueue(data)
        self.q1 , self.q2 = self.q2 , self.q1
        return data
        
    def traversing(self):
        self.q1.traverse()
```

```python
# Now lets implement Queue using Stack
class QueueS:
    def __init__(self):
        self.s1 = StackQ() # what this stack uses internally be it array or linkedlist or queue doesnt matter.
        self.s2 = StackQ()
    
    def enQue(self,data):
        self.s1.push(data)
    
    def isEmpty(self):
        return self.s1.isEmpty()
    
    def deQue(self):
        if (self.s1.isEmpty()):
            print("Queue underflow exception")
            return
        data = None
        while not self.s1.isEmpty():
            data = self.s1.pop()
            if(not self.s1.isEmpty()):
                self.s2.push(data)
        self.s1,self.s2 = self.s2,self.s1
        return data
    
    def traversing_q(self):
        self.s1.traversing()
```
